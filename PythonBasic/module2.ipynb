#dictionaries comprehension


```python
#num and its sq
sq={x:x*x for x in range(6)}
print(sq)
```

```python
#num and its sq of odd num
sq={x:x*x for x in range(6) if x%2==1}
print(sq)
```

```python
#Create a dictionary using comprehension from two sequences.
kys=['a','b','c']
vls=[1,2,3]
{i:j for (i,j) in zip(kys,vls)}
```

```python
#From a list of names, using comprehension create a dictionary which has the first character of each name as the key and the corresponding name as its value
n=eval(input("enter  name []"))
s={i[0]:i for i in n}
print(s)
```
